```python
from pyspark.mllib.recommendation import ALS, Rating

df = sc.textFile("/home/tushar/Desktop/redcarpet/ml-latest-small/ratings.csv")
```

```python
# showing all data of rdd
df.collect()
```

```python
ratings = df.map(lambda l: l.split(',')).map(lambda l: Rating(int(l[0]), int(l[1]), float(l[2])))
```

```python
# user input for which rank you wanna train 
rank = int(input("enter number:"))
numIterations = 10

# trainning model with ALs but some issues with spark 2.3.1 & Databricks too
model = ALS.train(ratings, rank, numIterations)

#getting testdata 
testdata = ratings.map(lambda p: (p[0], p[1]))
#making predications based on testdata
predictions = model.predictAll(testdata).map(lambda r: ((r[0], r[1]), r[2]))
ratesAndPreds = ratings.map(lambda r: ((r[0], r[1]), r[2])).join(predictions)
    
```

```python
# mean square error for this 
MSE = ratesAndPreds.map(lambda r: (r[1][0] - r[1][1])**2).mean()
print("Mean Square Error:",MSE)
```
